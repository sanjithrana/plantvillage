```python
# Get the list of files in the directory
image_dir_path = "/content/plantvillage dataset/color/Blueberry___healthy"
image_files = os.listdir(image_dir_path)

# Select one of the image files (e.g., the first one in the list)
# You might want to add a check here to ensure the directory is not empty
if image_files:
    image_filename = image_files[1]
    image_path = os.path.join(image_dir_path, image_filename)

    #read image
    img = mping.imread(image_path)

    print(img.shape)#display
    plt.imshow(img)
    plt.axis('off')
    plt.show()
else:
    print(f"No image files found in {image_dir_path}")
```

```python
image_size = 224
batch_size = 32
```

```python
#train and test
data_genn = ImageDataGenerator(
    rescale = 1./255,
    validation_split= 0.2
)
```

```python
#trian
train_genarator = data_genn.flow_from_directory(
    bas_dir,
    target_size = (image_size,image_size),
    batch_size= batch_size,
    subset = 'training',
    class_mode = 'categorical'

)
```

```python
validate_genarator = data_genn.flow_from_directory(
    bas_dir,
    target_size = (image_size,image_size),
    batch_size= batch_size,
    subset = 'training',
    class_mode = 'categorical'

)
```

```python
#cnn
model = models.Sequential()

model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(image_size,image_size,3)))
model.add(layers.MaxPool2D(pool_size =(2,2)))


model.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(layers.MaxPool2D(pool_size =(2,2)))

model.add(layers.Flatten())
model.add(Dense(256,activation='relu'))

model.add(layers.Dense(train_genarator.num_classes,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#model training

history = model.fit(train_genarator,steps_per_epoch=train_genarator.samples // batch_size,
                    epochs = 5,
                    validation_data = validate_genarator,
                    validation_steps = validate_genarator.samples // batch_size)
```

```python
#model_evalution
print('evalution model........')
val_loss,val_acc =model.evaluate(validate_genarator,steps = validate_genarator.samples  // batch_size)
print(f"validation : {val_acc * 100:.2f}%")
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
# bulid a predict model

def load_pross(image_path,target_size = (224,224)):
  #load image
  img = Image.open(image_path)
  #resize the image
  img = img.resize(target_size)
  #convert the img to a numpy
  img_array = np.array(img)
  #add batch dimension
  img_array = np.expand_dims(img_array,axis = 0)
  #scale the image value to [0,1]
  img_array = img_array.astype('float32') /255
  return img_array

#fun to predict the img
def predict_img(model,image_path,class_indices):
  preprocessed_img = load_pross(image_path)
  predicts = model.predict(preprocessed_img)
  predict_class_index = np.argmax(predicts,axis = 1)[0]
  predict_class_name = class_indices[predict_class_index]
  return predict_class_name
```

```python
class_indices = {v:k for k,v in train_genarator.class_indices.items()}
#create a mapping from class indices to class name
```
